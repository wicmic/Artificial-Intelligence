```python
import regex as re
import numpy as np
for index, row in df.iterrows():
    #print("----------------------")
    #print(row["METAR"])
    if not pd.isna(row["METAR"]):
        tokens = row["METAR"].split(' ')
        untergrenze = []
        error = 0
        weathercount = 0
        sicht = 0
        day = 0
        hour = 0
        minute = 0
        winddir = 0
        windspeed = 0
        windvariability = 0
        boe = 0
        sicht = 0
        ltemp = 0
        tautemp = 0
        ltemp = 0
        spread = 0  
        qnh = 0
        weathercount = 0
        lowestceiling = 0
        cloudcoverage = 0
            
        for token in tokens:
            if (re.match('NIL', token)):
                #print("ERROR: ", token)
                error =  1
            if (re.match('[0-9]{6}Z', token)):
                #print("Zeitstring: ", token)
                day = int(token[0:2])
                hour = int(token[2:4])
                minute = int(token[4:6])
                #print("Tag:",day," Stunde(UTC):",hour," Minute:", minute)
            if (re.match('[VRB0-9G]{5,9}KT', token)):
                #print("Windstring: ", token)
                wind = re.findall("([VRB\d]{5,5})", token)
                if (re.search('VRB', token)):
                    winddir = 0
                    windspeed = int(wind[0][3:5])
                    #print("Winddirection: ", '-none-')
                    #print("Windspeed: ", wind[0][3:5])
                wind = re.findall("([\d]{5,5})", token)
                if (len(wind) == 1):
                    winddir = int(wind[0][0:3])
                    windspeed = int(wind[0][3:5])
                    #print("Winddirection: ", wind[0][0:3])
                    #print("Windspeed: ", wind[0][3:5])
                if (re.search('G', token)):
                    boe = int(re.findall("G([\d]{2,3})KT", token)[0])
                    #print("Boe: ", re.findall("G([\d]{2,3})KT", token)[0])
            if (re.match('[0-9]{3}V[0-9]{3}', token)):
                windvariability = token
                #print("WindvariabilitÃ¤t: ", token)
            if (re.match('[\d]{3,4}$', token )):
                if (re.search('[\D]', token)):
                    pass
                else:
                    sicht = int(token)
            if (token == "CAVOK"):
                sicht = 10000
                #print("CAVOK: True -Sicht... kein Wetter....")
            if (re.match('[M0-9]{2,3}/[M0-9]{2,3}', token)):
                #print("Temperaturstring: ", token)
                temps = re.findall("[M\d]{1,3}", token)
                #print("T:",temps)
                if (len(temps) == 2):
                    ltemp = int(temps[0][1:3])
                    if (temps[0][0:1] == 'M'):
                        ltemp = -ltemp
                    #print("Lufttemperatur: ", ltemp)
                    tautemp = int(temps[1][1:3])
                    if (temps[1][0:1] == 'M'):
                        tautemp = -tautemp
                    #print("Taupunkttemperatur: ", tautemp)
                    spread = (ltemp-tautemp)*400
            if (re.match('Q[0-9]{3,4}[=]{0,1}', token)):
                #print("QNH: ", re.findall("Q([0-9]{3,4})", token)[0])
                qnh = int(re.findall("Q([0-9]{3,4})", token)[0])
            if (re.search('DR|TS|FZ|SG|PL|GR|GS|FG|SQ|RA|SH|DZ|BR|SN', token)):
                if (len(re.findall('NOSIG', token)) == 0):
                    #print("Wetter: ", token)
                    weathercount += 1

            if (re.findall('(FEW[\d]{3,3}|SCT[\d]{3,3}|BKN[\d]{3,3}|OVC[\d]{3,3})', token)):
                #print("Wolkenstring: ", token)
                if (re.match('FEW[\d]{3,3}', token)):
                    grad = 1
                if (re.match('SCT[\d]{3,3}', token)):
                    grad = 2
                if (re.match('BKN[\d]{3,3}', token)):
                    grad = 3
                if (re.match('OVC[\d]{3,3}', token)):
                    grad = 4
                untergrenze.append((grad, int(token[3:6])*100))
        #print("Untergrenzen:", untergrenze)
        
        #print("WetterzÃ¤hler: ", weathercount)
        ceilings = np.array(untergrenze)
        if(ceilings.size > 0):
            ceilings = np.insert(ceilings, 0, [(0,10000)], axis=0)
            #print(ceilings)
            lowestceiling = min(ceilings.T[1])
            cloudcoverage = sum(ceilings.T[0])
        else:
            lowestceiling = 10001
        #print("Niedrigste Wolkenuntergrenze: ", lowestceiling)
        #print("Wolkendichte: ", cloudcoverage)
        #print("Sicht: ", sicht)
        if (qnh < 700):
            error += 1
        df.at[index, "Error"] = error
        if (error > 0):
            pass
            #print ("ERROR in: ", row["METAR"])
        else:
            df.at[index, "Day"] = day
            df.at[index, "Hour"] = hour
            df.at[index, "Minute"] = minute
            df.at[index, "Winddir"] = winddir
            df.at[index, "Windspeed"] = windspeed
            #df.at[index, "Windvar"] = windvariability
            df.at[index, "Boe"] = boe
            df.at[index, "Sicht"] = sicht
            df.at[index, "L-Temp"] = ltemp
            df.at[index, "T-Temp"] = tautemp
            df.at[index, "L-Temp"] = ltemp
            df.at[index, "Spread"] = spread
            df.at[index, "QNH"] = qnh
            df.at[index, "Wettercount"] = weathercount
            df.at[index, "Ceiling"] = lowestceiling
            df.at[index, "Cloudcoverage"] = cloudcoverage
            
            score = weathercount * 2
            if (windspeed >= 17):
                score += 1
            if (windspeed >= 25 or boe > 20):
                score += 1
            if (spread < 500):
                score += 1
            if (lowestceiling < 800):
                score += 1
            if (sicht < 3000):
                score += 1
            if (sicht < 2000):
                score += 1
                
            df.at[index, "Wetterscore"] = score
            
            
```

```python
df = df[df['Error'] == 0]
```

```python
df = df[['Year', 'Month','Day', 'Hour', 'Wetterscore', 'QNH', 'METAR']]
```

```python
df['wcsma6'] = df['Wetterscore'].rolling(window=6).mean()
df['wcsma12'] = df['Wetterscore'].rolling(window=12).mean()
df['wcsma24'] = df['Wetterscore'].rolling(window=24).mean()
df['wcsma48'] = df['Wetterscore'].rolling(window=48).mean()
df['wcsma72'] = df['Wetterscore'].rolling(window=72).mean()

df['dwcsma'] = np.gradient(df['Wetterscore'], 1) 
df['dwcsma6'] = np.gradient(df['Wetterscore'].rolling(window=6).mean(), 1) 
df['dwcsma12'] = np.gradient(df['Wetterscore'].rolling(window=12).mean(), 1)
df['dwcsma24'] = np.gradient(df['Wetterscore'].rolling(window=24).mean(), 1)
df['dwcsma48'] = np.gradient(df['Wetterscore'].rolling(window=48).mean(), 1)
df['dwcsma72'] = np.gradient(df['Wetterscore'].rolling(window=72).mean(), 1)

df['QNH6'] = df['QNH'].rolling(window=6).mean()
df['QNH12'] = df['QNH'].rolling(window=12).mean()
df['dQNH'] = np.gradient(df['QNH'], 1)
df['dQNH6'] = np.gradient(df['QNH'].rolling(window=6).mean(), 1)
df['dQNH12'] = np.gradient(df['QNH'].rolling(window=12).mean(), 1)
```

```python
import pylunar
import math
mi = pylunar.MoonInfo((48, 30, 00), (10, 0, 0))

for index, row in df.iterrows():
        #print(row["Year"],row["Month"],row["Day"])
        mi.update((int(row["Year"]), int(row["Month"]), int(row["Day"]), int(row["Hour"]), 0, 0))
        mondphase = mi.fractional_phase()
        mondtag = mi.age()
        #print(row["Year"], row["Month"], row["Day"], row["Hour"], "Ph:", mondphase, " age: ", mondtag)
        df.at[index, "Mondphase"] = mondphase
        df.at[index, "Mondtag"] = mondtag
        if (mondtag <= 14.71):
            df.at[index, "ZuOderAbnehmend"] = 'Z'
        else:
            df.at[index, "ZuOderAbnehmend"] = 'A'
```

```python
df['Mondquadrant'] = np.digitize(df['Mondtag'], np.linspace(0,max(df['Mondtag']),5), right=True)
df['Mondoktant'] = np.digitize(df['Mondtag'], np.linspace(0,max(df['Mondtag']),9), right=True)
df['Mondhextant'] = np.digitize(df['Mondtag'], np.linspace(0,max(df['Mondtag']),17), right=True)
df
```

```python
# sampling rate
sr = len(df['Mondtag'])
# sampling interval
ts = 1.0/sr
t = np.arange(0,1,ts)

freq = 1.
x = df['wcsma72']


plt.figure(figsize = (28, 6))
plt.plot(t, x, 'r')
plt.ylabel('Amplitude')

plt.show()
```

```python
from numpy.fft import fft, ifft

X = fft(x)
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 

plt.figure(figsize = (12, 6))
plt.subplot(121)

plt.stem(freq, np.abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')
plt.xlim(0, 10)

plt.subplot(122)
plt.plot(t, ifft(X), 'r')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```
